```python
"""itertools"""
#Itertools is a module in Python, it is used to iterate over data structures that can be stepped over using a 
#for-loop. Such data structures are also known as iterables. This module works as a fast, memory-efficient tool
#that is used either by themselves or in combination to form iterator algebra.



#repeat: infinite loop unless specified a value

"""product"""
#as name implies multplies the items.lists,tups
#is used to find the cartesian product from the given iterator, output is lexicographic ordered
from itertools import product
a = "Krishna",
b = "bannurkar","Katte"

prod = product(a,b, repeat = 2)
print(tuple(prod))

"""permutations"""
#permutation relates to the act of arranging all the members of a set into some sequence or orde
from itertools import permutations
a = "Krishna"
perm = permutations(a)
#for alphabets in list(perm):
#    print(alphabets)
b = (1,2,3)
perm2 = permutations(b)
print(tuple(perm2))


"""combinations"""
#combination is a way of selecting items from a collection, 
#such that (unlike permutations) the order of selection does not matter.
from itertools import combinations, combinations_with_replacement
a = "Krishna"
comb = combinations(a,1)
for alphabets in list(comb):
    print(alphabets)
b = (1,2,3)
comb = combinations(b,2)
print(tuple(comb))
comb_wr = combinations_with_replacement(b,2)
print(tuple(comb_wr))


"""accumulate"""
from itertools import accumulate
import operator

#it accumulates and add to the next item by default we can specify an operator


a = [1,2,3,4,5,6]
acc = accumulate(a)
print(list(acc))
print(type(a))

accu = accumulate(a, func=operator.mul)
print(list(accu))

b = [1,3,6,3,7,2,9]
accu = accumulate(b, func=max)
print(b)
print(list(accu))




"""groupby"""
import itertools
from itertools import groupby
persons = [{"Name" : "Krishna" , "age" : "26"}, {"Name" : "Nikhil", "age": "30"}, 
           {"Name" : "Ashu" , "age" : "26"}, {"Name" : "Vijay", "age": "30"}]

key_f = lambda x : x["age"]
for key , group in  itertools.groupby(persons , key_f):
    print(key + " :", list(group))
    

L = [("a", 1), ("a", 2), ("b", 3), ("b", 4)]
  
# Key function
key_func = lambda x:  [0]
  
for key, group in itertools.groupby(L, key_func):
    print(key + " :", list(group))
```
